```python
import pandas as pd
import os

file_paths = [
    "/content/drive/MyDrive/일조시간관측데이터/OBS_ASOS_DD_20251107174059(20150101-20170101).csv",
    "/content/drive/MyDrive/일조시간관측데이터/OBS_ASOS_DD_20251107175904(20170101-20190102).csv",
    "/content/drive/MyDrive/일조시간관측데이터/OBS_ASOS_DD_20251107180309(20190103-20210103).csv",
    "/content/drive/MyDrive/일조시간관측데이터/OBS_ASOS_DD_20251107180554(20210104-20230104).csv",
    "/content/drive/MyDrive/일조시간관측데이터/OBS_ASOS_DD_20251107181409(20230105-20250105).csv",
    "/content/drive/MyDrive/일조시간관측데이터/OBS_ASOS_DD_20251107181910(20250106-20251106).csv"
]

# List to hold individual DataFrames
dfs = []

# Loop through each file path, read the CSV, and append to the list
for file_path in file_paths:
    try:
        df = pd.read_csv(file_path, encoding='euc-kr') # Assuming 'euc-kr' encoding for Korean data
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Concatenate all DataFrames into a single DataFrame
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    print(f"Successfully combined {len(dfs)} files into a single DataFrame.")
    print("Combined DataFrame head:")
    display(combined_df.head())
    print("Combined DataFrame info:")
    combined_df.info()
else:
    print("No DataFrames were loaded.")
```

```python
import pandas as pd

file_path = "/content/drive/MyDrive/일조시간관측데이터/META_관측지점정보_20251107162433.csv"

try:
    df = pd.read_csv(file_path, encoding='euc-kr') # Assuming 'euc-kr' encoding for Korean data
    print(f"Successfully loaded {file_path}.")
    print("DataFrame head:")
    display(df.head())
    print("DataFrame info:")
    df.info()
except Exception as e:
    print(f"Error reading {file_path}: {e}")
```

```python
df_filtered = df[df['종료일'].isnull()].copy()
print("DataFrame after removing rows with non-null '종료일':")
display(df_filtered.head())
print("Filtered DataFrame info:")
df_filtered.info()
```

```python
import pandas as pd

# 1. 메타 데이터(df_filtered)에서 '지점'을 기준으로 중복 제거
# 위도, 경도 정보만 필요하므로 필요한 컬럼만 선택 후 중복을 제거합니다.
# 이 과정은 메타 데이터만 다루므로 원본 데이터의 '일시'는 영향받지 않습니다.
df_meta_unique = df_filtered[['지점', '위도', '경도']].drop_duplicates(subset=['지점'])

# 2. 원본 데이터(combined_df)에 병합 (Left Merge)
combined_df_merged = pd.merge(
    combined_df,      # 기준이 되는 데이터 ('일시' 포함됨)
    df_meta_unique,   # 중복이 제거된 메타 데이터
    on='지점',        # '지점' 컬럼을 기준으로 병합
    how='left'        # combined_df의 모든 데이터를 유지
)

# 3. 결과 확인
print(f"병합 전 combined_df 행 개수: {len(combined_df)}")
print(f"병합 후 combined_df_merged 행 개수: {len(combined_df_merged)}")

# 혹시 모를 병합 후 중복 확인 (선택 사항)
if len(combined_df) != len(combined_df_merged):
    print("경고: 병합 후 행 개수가 일치하지 않습니다. 메타 데이터를 다시 확인해주세요.")
else:
    print("성공: 행 개수 변화 없이 안전하게 병합되었습니다.")

display(combined_df_merged.head())
combined_df_merged.info()
```

```python
combined_df_merged['위도'] = combined_df_merged['위도'].fillna(35.8782)
combined_df_merged['경도'] = combined_df_merged['경도'].fillna(128.6522)

print("결측값 채운 후 combined_df_merged의 상위 5개 행:")
display(combined_df_merged.head())

print("결측값 채운 후 combined_df_merged의 결측치 확인:")
display(combined_df_merged.isnull().sum())
```

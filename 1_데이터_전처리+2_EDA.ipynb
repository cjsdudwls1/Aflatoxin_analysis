# 데이터 결합

```python
import pandas as pd

df_X = pd.read_pickle('/content/drive/MyDrive/251108다시분석데이터/df_통합_LIMS_기상정보_일조포함_결합_60일_gzip_X.pkl', compression='gzip')
df_y = pd.read_pickle('/content/drive/MyDrive/251108다시분석데이터/df_통합_LIMS_기상정보_일조포함_결합_60일_gzip_y.pkl', compression='gzip')

print("df_X head:")
display(df_X.head())

print("\ndf_y head:")
display(df_y.head())
```

# 불필요데이터 삭제

```python
# 제거 전 df_combined의 행 수
initial_rows = df_combined.shape[0]

# 'JDGMNT_WORD_NAME' 칼럼에서 '상기실험확인함' 값을 갖는 행을 식별합니다.
mask_to_remove = df_combined['JDGMNT_WORD_NAME'] == '상기실험확인함'

# 마스크를 반전시켜 '상기실험확인함'이 아닌 행만 선택하여 데이터프레임을 업데이트합니다.
df_combined = df_combined[~mask_to_remove]

# 제거 후 df_combined의 행 수
final_rows = df_combined.shape[0]

# 제거된 행의 수 계산
removed_rows_count = initial_rows - final_rows

print(f"총 {removed_rows_count}개의 행이 제거되었습니다.")
print(f"제거 후 df_combined의 새로운 shape: {df_combined.shape}")

print("\ndf_combined head after filtering:")
display(df_combined.head())
```

```python
import re

def fix_nan_auto_group(df):

    '''
    시계열자료 채우기 로직
    '''
    result = df.copy()

    # 컬럼 이름에서 그룹 접두사(e.g., 'wtf', 'vsf')를 자동으로 추출
    # '_01' 과 같은 부분을 제외하고 고유한 그룹 이름을 찾음
    groups = {re.findall(r'(\w+)_', col)[0] for col in df.columns if '_' in col}

    # 각 그룹별로 NaN 값 채우기
    for group_name in groups:
        group_cols = [col for col in df.columns if col.startswith(group_name)]
        # 그룹 내에서 뒤쪽 값으로 먼저 채운 후 앞쪽 값으로 채우기
        result[group_cols] = result[group_cols].bfill(axis=1).ffill(axis=1)

    # 그룹 내에서 채워지지 않은 NaN(행의 양끝)을 0으로 채우기
    result = result.fillna(0)

    return result
```
